# Markov-switching multifractal model

To replicate the results for the Markov-switching multifractal model, execute this notebook using a [Julia]((https://julialang.org/)) (>= 1.6.1) kernel.

```julia
using DrWatson
@quickactivate "Zila_Kukacka_2023"

include(srcdir("fcn_utils.jl"))
include(srcdir("fcn_plots.jl"));
```

```julia
folder_bsw = "markov256state_bsw_lux2022_lux2022_rep96_obs6750_burn200_inits1_sim100_iter2000_fw2012overlap_blocksize250_bootsize5000_blockcount1000"
folder_fsw = "markov256state_fsw_lux2022_lux2022_rep96_obs6750_burn200_inits1_sim100_iter2000_fw2012overlap_blocksize250_bootsize5000_blockcount1000";
```

```julia
mod_set  = foldername_mod_set_retrieve(folder_bsw)
bench = load_bench(folder_bsw);
```

```julia
results_bsw = load_results(folder_bsw)
results_fsw = load_results(folder_fsw);
```

```julia
display("Markov-switching multifractal: BSME")
plot_rmse(results_bsw, mod_set, folder_bsw, bench=bench, bench_names=["CHL4", "CHL15", "FW9"], ylimits=(0.0, 0.2), show_plot=true)
display("Markov-switching multifractal: FSMS")
plot_rmse(results_fsw, mod_set, folder_fsw, bench=bench, bench_names=["CHL4", "CHL15", "FW9"], ylimits=(0.0, 0.2), show_plot=true);
```

```julia
display("Markov-switching multifractal: BSME")
plot_evolution_bias(results_bsw, mod_set, folder_bsw, "bsw", (0.0, 0.15), show_plot=true)
display("Markov-switching multifractal: FSMS")
plot_evolution_bias(results_fsw, mod_set, folder_fsw, "fsw", (0.0, 0.15), show_plot=true);
```
